Goal is to compare change-point convergence across thresholds and values of epsilon.  

```python
sims = 50
M = 10
nu = int(1e4)
T = int(1e6)
mu1 = 1
thresholds = [100,200,300,400,500]
epsilons = [0.01,0.05,0.1,0.25,0.5]
```

```python
import focus_implementation as focus
import numpy as np

edd_means = np.zeros((len(epsilons),len(thresholds)))
edd_std = np.zeros((len(epsilons),len(thresholds)))
for idx, epsilon in enumerate(epsilons):
    for idx2, threshold in enumerate(thresholds):
        correct_detections = 0
        delays = []
        while correct_detections < sims:
            data = focus.generate_data(M,T,nu,mu1)
            detection_time, _ = focus.eps_focus(data,epsilon,threshold)
            if detection_time>nu:
                correct_detections = correct_detections + 1
                delays.append(detection_time-nu)
        print(f"epsilon: {epsilon}")
        print(f"threshold: {threshold}")
        print(sum(delays)/correct_detections)
        edd_means[idx][idx2] = sum(delays)/correct_detections
        edd_std[idx][idx2] = np.std(delays)
```
